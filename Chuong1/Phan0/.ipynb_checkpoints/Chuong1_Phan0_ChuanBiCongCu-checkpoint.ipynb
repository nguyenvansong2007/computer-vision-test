```python
# đọc  và hiển thị nhiều ảnh trong một khung hình
img_muaxuan = cv2.imread('img/muaxuan.png')
```

```python
img_muaha = cv2.imread('img/muaha.jpg')
```

```python
img_muathu = cv2.imread('img/muathu.jpeg')
```

```python
img_muadong = cv2.imread('img/muadong.jpg')
```

```python
# hiển thị bốn ảnh vào trong một khung hình
plt.figure(figsize=(12,8))

plt.subplot(2,2,1) # hiển thị mùa xuân
plt.imshow(img_muaxuan,cmap='gray')
plt.title("Ảnh mùa xuân")
plt.grid(ls='--')

plt.subplot(2,2,2) # hiển thị mùa hạ
plt.imshow(img_muaha,cmap='gray')
plt.title("Ảnh mùa hạ")
plt.grid(ls='--')

plt.subplot(2,2,3) # hiển thị mùa thu
plt.imshow(img_muathu,cmap='gray')
plt.title("Ảnh mùa thu")
plt.grid(ls='--')

plt.subplot(2,2,4) # hiển thị mùa đông
plt.imshow(img_muadong,cmap='gray')
plt.title("Ảnh mùa đông")
plt.grid(ls='--')
```

```python
img_bro1 = cv2.imread('img/bro1.jpg')
img_bro2 = cv2.imread('img/bro2.jpg')
img_bro3 = cv2.imread('img/bro3.jpg')
img_bro = cv2.imread('img/lac.jpg')
```

```python
plt.figure(figsize=(12,8))

plt.subplot(1,4,1) # hiển thị mùa xuân
plt.imshow(img_bro1,cmap='gray')
plt.title("Ảnh ngủ gật")
plt.grid(ls='--')

plt.subplot(1,4,2) # hiển thị mùa xuân
plt.imshow(img_bro,cmap='gray')
plt.title("Người nghiện lâu năm")
plt.grid(ls='--')

plt.subplot(1,4,3) # hiển thị mùa xuân
plt.imshow(img_bro2,cmap='gray')
plt.title("không biết để cap gì")
plt.grid(ls='--')

plt.subplot(1,4,4) # hiển thị mùa xuân
plt.imshow(img_bro3,cmap='gray')
plt.title("bò sữa")
plt.grid(ls='--')

```
