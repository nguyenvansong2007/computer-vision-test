```python
#Khai báo sử dụng thư viện:
import numpy as np
import matplotlib.pyplot as plt
import cv2
print('Version OpenCV:', cv2.__version__)
```

## THỰC HÀNH 4.1
---

1. Đọc và hiển thị ảnh Thuchanh4_1.jpeg ở dạng ảnh xám
2. Phân vùng ảnh sử dụng phương pháp phân ngưỡng toàn cục (Hãy xác định ngưỡng phù hợp để đạt kết quả tốt nhất)
3. Phân vùng ảnh sử dụng phương pháp phân ngưỡng thích nghi (Hãy xác định các tham số phù hợp để đạt kết quả tốt nhất)

<img src='images/Thuchanh4_1.jpeg' width='300px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
#Đọc ảnh hạt thóc ở chế độ ảnh xám:
img_thoc = cv2.imread('images/Thuchanh4_1.jpeg', 0)

#Hiển thị ảnh gốc:
plt.imshow(img_thoc, cmap='gray')
plt.show()
```

```python
#Hiển thị histogram của ảnh để xác định ngưỡng phù hợp:
plt.hist(img_thoc.ravel(), 256, [0,255])
plt.show()
```

```python
#Lựa chọn ngưỡng và thực hiện cắt ngưỡng cục bộ: T = 95 - Chọn ngưỡng làm sao vẫn nhìn thấy hạt thóc
(T, thresh1) = cv2.threshold(img_thoc, 95, 255, cv2.THRESH_BINARY) #Ngưỡng khá quan trọng (200-255)

#Hiển thị kết quả:
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(img_thoc, cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1,2,2)
plt.imshow(thresh1, cmap='gray')
plt.title('Ảnh thóc cắt ngưỡng toàn cục')
plt.show()
```

```python
#Thực hiện phân ngưỡng thích nghi
#Phương pháp mean:
thoc1 = cv2.adaptiveThreshold(img_thoc, 255,
                                cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY,
                                55, 10)
#Phương pháp Gaussian
thoc2 = cv2.adaptiveThreshold(img_thoc, 255,
                               cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY_INV,
                               21, 10)

#Hiển thị kết quả:
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.imshow(img_thoc, cmap='gray')
plt.title('Ảnh gốc')

plt.subplot(1,3,2)
plt.imshow(thoc2, cmap='gray')
plt.title('Cắt thóc ngưỡng thích nghi Gaussian')


plt.subplot(1,3,3)
plt.imshow(thoc1, cmap='gray')
plt.title('Cắt thóc ngưỡng thích nghi Mean')

plt.show()
```

# THỰC HÀNH 4.2
---
1. Đọc và hiển thị ảnh Thuchanh4_2.jpeg ở dạng ảnh màu.
2. Sử dụng phương pháp Canny với các tham số phù hợp phát hiện biên của các đối tượng trong ảnh.
3. Đếm số lượng bóng bay trong bức ảnh với OpenCV

<img src='images/Thuchanh4_2.jpeg' width='400px'>


```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
bong_bay = cv2.imread('images/Thuchanh4_2.jpeg')
bong_bay_1 = cv2.cvtColor(bong_bay, cv2.COLOR_BGR2RGB) 


plt.figure(figsize=(10,8))
plt.imshow(bong_bay_1)
plt.show()
```

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
#làm trơn vs gaussian
img_b = cv2.GaussianBlur(bong_bay_1,(5,5),0)

#Sử dụng phương thức canny:
img_canny1 = cv2.Canny(img_b, 50,100)
img_canny2 = cv2.Canny(img_b, 80,170)

#Hiển thị kết quả:
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)

plt.imshow(img_b,cmap='gray')
plt.title('Ảnh gốc đã làm mờ')

plt.subplot(1, 3, 2)
plt.imshow(img_canny1,cmap='gray')
plt.title('Biên với Canny1 ')

plt.subplot(1, 3, 3)
plt.imshow(img_canny2,cmap='gray')
plt.title('Biên với Canny2')

plt.show()



```

```python
#Bước 1: Đọc ảnh ở chế độ ảnh màu:
image = cv2.imread('images/Thuchanh4_2.jpeg')

#Hiển thị ảnh gốc ở dạng xám:
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray');
plt.title('Ảnh gốc')

#Bước 2: Phát hiện biên với Canny:
#Thực hiện làm mịn ảnh với Gausian:
blur = cv2.GaussianBlur(gray, (11,11), 0)

#Phát hiện biên với Canny
canny = cv2.Canny(blur, 30, 150, 3)
plt.subplot(2,2,2)
plt.imshow(canny, cmap='gray')
plt.title('Ảnh biên với Canny')


#Bước 3: Làm đậm đường biên:
dilated = cv2.dilate(canny, (1,1), iterations = 2)
plt.subplot(2,2,3)
plt.imshow(dilated, cmap='gray')
plt.title('Ảnh làm đậm đường biên')

#Bước 4: Lấy countour của ảnh và đếm số lượng:
(cnt, heirarchy) = cv2.findContours(dilated.copy(), 
                                    cv2.RETR_EXTERNAL, 
                                    cv2.CHAIN_APPROX_NONE)

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (200,0,0), 2)

plt.subplot(2,2,4)
plt.imshow(rgb)
plt.title('Ảnh chồng lấp countour')
plt.show()

print('Số lượng đồng xu trong ảnh là: ', len(cnt))
```

```python
#Viết hàm hiển thị các pixel điểm ảnh và màu sắc
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data  
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

# THỰC HÀNH 4.3
---
1. Đọc và hiển thị ảnh Thuchanh4_3.jpeg ở dạng ảnh màu.
2. Sử dụng phương pháp phân vùng ảnh dựa trên kỹ thuật phân cụm Kmeans để phân ảnh ra thành:
    * 1 cụm
    * 2 cụm
    * 3 cụm
    
* Hiển thị kết quả: ảnh 1 cụm - ảnh 2 cụm - ảnh 3 cụm

<img src='images/Thuchanh4_3.jpeg' width='400px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}


img_samac = cv2.imread('images/Thuchanh4_3.jpeg',1)
img_samac = cv2.cvtColor(img_samac,cv2.COLOR_BGR2RGB)
plt.imshow(img_samac,cmap='gray')
plt.title('ảnh gốc sa mạc')
plt.show()
```

```python
#Biến đổi dữ liệu ảnh về dạng phù hợp với mô hình:
#Scaling data đưa dữ liệu từ khoảng [0--> 255] về khoảng [0-1]
data=img_samac/255.0

#Đưa dữ liệu ảnh màu từ mảng 3 chiều về ma trận:
data = data.reshape(-1, 3)

data.shape
```

```python
#Viết hàm hiển thị các pixel điểm ảnh và màu sắc
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data  
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

```python
#Sử dụng thư viện học máy Scikit - Learn:
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans

#Phân thành 1 cụm:
kmeans_images_1 = KMeans(n_clusters=1)
kmeans_images_1.fit(data)

#Minh họa kết quả phân cụm các pixel điểm ảnh:
new_colors_1 = kmeans_images_1.cluster_centers_[kmeans_images_1.predict(data)]
plot_pixels(data, colors=new_colors_1,
            title="Reduced color space with Kmeans")
```

```python
#Sử dụng thư viện học máy Scikit - Learn:
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans

#Phân thành 2 cụm:
kmeans_images_2 = KMeans(n_clusters=2)
kmeans_images_2.fit(data)

#Minh họa kết quả phân cụm các pixel điểm ảnh:
new_colors_2 = kmeans_images_2.cluster_centers_[kmeans_images_2.predict(data)]
plot_pixels(data, colors=new_colors_2,
            title="Reduced color space with Kmeans")
```

```python
#Sử dụng thư viện học máy Scikit - Learn:
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans

#Phân thành 3 cụm:
kmeans_images_3 = KMeans(n_clusters=3)
kmeans_images_3.fit(data)

#Minh họa kết quả phân cụm các pixel điểm ảnh:
new_colors_3 = kmeans_images_3.cluster_centers_[kmeans_images_3.predict(data)]
plot_pixels(data, colors=new_colors_3,
            title="Reduced color space with Kmeans")
```

```python
samac_recolored_1 = new_colors_1.reshape(img_samac.shape)
samac_recolored_2 = new_colors_2.reshape(img_samac.shape)
samac_recolored_3 = new_colors_3.reshape(img_samac.shape)



#Hiển thị kết quả:
plt.figure(figsize=(16,6))
plt.subplot(1,4,1)
plt.imshow(img_samac)
plt.title('Ảnh gốc')

plt.subplot(1,4,2)
plt.imshow(samac_recolored_1)
plt.title('Kmeans 1 cụm')

plt.subplot(1,4,3)
plt.imshow(samac_recolored_2)
plt.title('Kmeans 2 cụm')

plt.subplot(1,4,4)
plt.imshow(samac_recolored_3)
plt.title('Kmeans 3 cụm')

plt.show()
```
